## Load the chipotle dataset from SQL. 
### Create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
chipotle_db_url = f'mysql+pymysql://{env.user}:{env.password}@{env.host}/chipotle'
```

```python
chipotle_query = 'select * from orders'
```

```python
chipotle_df = pd.read_sql(chipotle_query, chipotle_db_url)
```

```python
chipotle_df
```

```python
chipotle_df.item_price = chipotle_df.item_price.str.strip('$').astype(float)
```

```python
top_4 = chipotle_df.groupby('item_name').quantity.sum().nlargest(4)
top_4
```

```python
four_pop= chipotle_df.groupby('item_name').quantity.agg(['sum']).sort_values(by='sum', ascending = False).head(4)
four_pop
```

```python
all_prices = chipotle_df.groupby('item_name').item_price.sum()

#take a look
all_prices.head()
```

```python
merged_df = pd.merge(top_4, all_prices, on='item_name', how='inner')
merged_df = merged_df.reset_index()
merged_df
```

```python
#create the plot
sns.barplot(data= merged_df, y='item_name', x='item_price')

#add title and label
plt.title('Four Most Popular Chipotle Items')
plt.xlabel('Total Revenue')
plt.show()
```
