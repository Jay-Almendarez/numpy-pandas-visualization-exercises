### 3. Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x=np.arange(150)
y=x**.5
plt.subplot(2,2,1)
plt.plot(x,y, c='red')
plt.title('Square Roots from 0 to 150')
plt.xlabel('Numbers to be square rooted')
plt.ylabel('Square root of x')


x=np.arange(-15,15)
y=x**3
plt.subplot(2,2,2)
plt.plot(x,y, c ='green')
plt.title('Range from 0 to 50 Cubed')
plt.xlabel('Non-suspecting numbers')
plt.ylabel('Cubed result')

x=np.arange(10)
y=2**x
plt.subplot(2,2,3)
plt.plot(x,y, c ='cyan')
plt.title('2 to the power of X')
plt.xlabel('Extent of Exponentiation')
plt.ylabel('Exponentiation of 2')

x=np.arange(10)
y=1/(x+1)
plt.subplot(2,2,4)
plt.plot(x,y, c ='purple')
plt.title('Ever Decreasing Fractions')
plt.xlabel('Denominator')
plt.ylabel('Fraction dictated')

plt.suptitle('Combined figures')
plt.subplots_adjust(wspace=.75, hspace=.75)
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = np.arange(0,10)
root_of_x = x**.5
x_cubed = x**3
two_to_x = 2**x
fraction_of_x = 1/(x+1)

plt.figure(figsize=(10, 8))

plt.plot(x, root_of_x, c='blue', label='$âˆšx$')
plt.plot(x, x_cubed, c='red', label='$x^3$')
plt.plot(x, two_to_x, c='green', label='$2^x$')
plt.plot(x, fraction_of_x, c='orange', label='$1/(x+1)$')

plt.xlim(0,8)
plt.ylim(0,20)
plt.grid(ls='-')

plt.legend(loc='upper left')
plt.title('Equations of X')
plt.xlabel('Elements')
plt.ylabel('Operations')
```
