### Curves to graph
- y=0n+1 and label the curve "O(1)"
- y=log(n) and label the curve "O(log n)"
- y=n and label the curve "O(n)"
- y=nâˆ—log(n) and label it "O(n log n)"
- y=n^2 and label it "O(n^2)"
- y=2^n and label it "O(2^n)"
- y=n! and label it "O(n!)"
- y=n^n and label it "O(n^n)"

```python
n=np.arange(1,30)
a=(n * 0)+1
b=np.log(n)
c=n
d=n * np.log(n)
e=n^2
f=2^n
g=n * (n - 1)
h=n^n

plt.plot(n, a, c='red', label='$O(1)$')
plt.plot(n, b, c='orange', label='$O(log n)$')
plt.plot(n, c, c='yellow', label='$O(n)$')
plt.plot(n, d, c='green', label='$O(n log n)$')
plt.plot(n, e, c='blue', label='$O(n^2)$')
plt.plot(n, f, c='cyan', label='$O(2^n)$')
plt.plot(n, g, c='violet', label='$O(n!)$')
plt.plot(n, h, c='black', label='$O(n^n)$')

plt.xlim(1,10)
plt.ylim(0,15)
plt.grid(ls='--')

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc='upper left')
```
